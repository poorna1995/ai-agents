```python
import requests
import json

def get_metadata(url):
    """
    Recursively fetch metadata from a GitHub repository's contents API and retrieve text from files.

    Parameters:
    - url (str): GitHub repository tree URL.

    Returns:
    - list of dict: List of file metadata with their content.
    """
    parts = url.split('/')
    owner = parts[3]
    repo = parts[4]
    branch = parts[6]
    path = '/'.join(parts[7:])

    # GitHub API URL for the repository contents
    api_url = f"https://api.github.com/repos/{owner}/{repo}/contents/{path}?ref={branch}"
    response = requests.get(api_url)

    if response.status_code == 200:
        json_output = response.json()
        list_metadata = []

        if isinstance(json_output, dict):
            json_output = [json_output]  # Ensure single file response is treated as a list

        for meta_data in json_output:
            if meta_data["type"] == "dir":
                # Recursively fetch contents from subdirectories
                sub_metadata = get_metadata(meta_data["html_url"].replace("https://github.com", "https://api.github.com/repos").replace("/tree/", "/contents/"))
                list_metadata.extend(sub_metadata)
            elif meta_data["type"] == "file":
                file_content = get_file_content(meta_data["download_url"])
                list_metadata.append({
                    "name": meta_data["name"],
                    "path": meta_data["path"],
                    "content": file_content
                })
        return list_metadata
    else:
        print(f"Failed to fetch metadata. Status code: {response.status_code}")
        return []

def get_file_content(file_url):
    """
    Fetch the raw content of a file from its download URL.

    Parameters:
    - file_url (str): Raw file download URL.

    Returns:
    - str: File content as text.
    """
    response = requests.get(file_url)
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to fetch file content. Status code: {response.status_code}")
        return ""

# Example URL
url = "https://github.com/duplocloud/docs/tree/main/getting-started-1/application-focussed-interface"
metadata = get_metadata(url)

# Print metadata and content
for file_data in metadata:
    print(f"File: {file_data['name']}")
    print(f"Path: {file_data['path']}")
    print("Content:")
    print(file_data['content'])
    print("=" * 80)
```
